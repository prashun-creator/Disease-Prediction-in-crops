```python
df['LW'] = df['LW'].apply(lambda x: 1 if x > 0.01 else 0)
```

```python
dfhat = pd.merge(df.groupby('date', sort=False)['date', 'TC'].mean(), df.groupby('date', sort=False)['date', 'LW'].sum(), on='date')
```

```python
dfhat = pd.merge(dfhat, df.groupby('date', sort=False)['date', 'PLV2'].sum(), on='date')
```

```python
def get_risk(x):
  if x.TC < 5:
    return 'Low'
  elif 18 > x.TC >= 5:
    if x.LW < 14:
      return 'Low'
    else:
      return 'Medium'
  elif 25 > x.TC >= 18:
    if x.LW < 6:
      return 'Low'
    elif 12 > x.LW >= 6:
      return 'Medium'
    else:
      return 'High'
  elif x.TC >= 30:
    return 'Low'
  else:
    if x.LW >= 10:
      return 'Medium'
    else:
      return 'Low'

```

```python
dfhat['Risktw'] = dfhat.apply(lambda row: get_risk(row), axis=1)
```

```python
dfhat
```

```python
dfhat['Risktw'].value_counts()
```

```python
dfhat['rain_int'] = dfhat.loc[:,'PLV2'].rolling(window=10).mean()
```

```python
def risk_rain(x):
  if 0.28 <= x.rain_int < 3 or x.rain_int >= 25:
    return 'Low'
  elif   3 <= x.rain_int < 7:
    return 'Medium'
  elif 7 <= x.rain_int < 25:
    return "High"
  else:
    return 'Low'
```

```python
dfhat['Risk_rain'] = dfhat.apply(lambda row: risk_rain(row), axis=1)
```

```python
dfhat
```

```python
def final_risk(x):
  dic = {"Low":1, 'Medium':2, 'High':3}
  return max(x.Risktw, x.Risk_rain, key=lambda x: dic[x])
```

```python
# dfhat.apply(lambda row: final_risk(row), axis=1)
```

```python
dfhat['final_risk'] = dfhat.apply(lambda row: final_risk(row), axis=1)
```

```python
dfhat
```

```python
dfhat['risk_factor'] = dfhat['final_risk'].apply(lambda x: {'Low':0, 'Medium':1, 'High':2}[x])
```

```python
dfhat['roll_risk'] = dfhat.loc[:,'risk_factor'].rolling(window=10).sum()
```

```python
ls = list(dfhat['roll_risk'])
```

```python
def spray(ls):
  pat = ['No']*len(ls)
  i = 0
  while i < len(ls):
    if i % 10 == 9:
      if ls[i] > 8:
        pat[i] = 'Yes'
        i += 10
      else:
        pat[i+5] = 'Yes'
        i += 5
    else:
      i += 1
  return pat
```

```python
set(spray(ls))
```

```python
dfhat['Spray'] = spray(ls)   #new column for spray in dataset
```

```python

from pymongo import MongoClient
import datetime


def get_data(crop_name, start_date, end_date, sensor_list, db_url):
  client = MongoClient(db_url)
  db = client.database
  lst = []
  customers, farms, devices   = db['customer'], db['farm'], db['device']
  for farm in farms.find({'name': crop_name}):
    query = {'deviceId':farm['deviceId'], 'datetime': {'$gte': datetime.datetime(start_date), '$lt': datetime.datetime(end_date)}}
    filter = {id: 1 for id in sensor_list}
    lst.extend(list(devices.find(query, filter)))
  return lst
      

```
